```python

from scipy import stats

import numpy as np
# calculate the Pearson's correlation between two variables
from numpy import mean
from numpy import std
from numpy import cov
from numpy.random import randn
from numpy.random import seed
from matplotlib import pyplot as plt
import seaborn as sns

from scipy.stats import pearsonr
# seed random number generator
seed(1)

# prepare data
data1 = np.random.uniform(low=55, high=100, size=100)
data2 = np.random.uniform(low=45, high=100, size=100)


# calculate covariance matrix
covariance = cov(data1, data2)

# calculate Pearson's correlation
corr, _ = pearsonr(data1, data2)

# plot
plt.scatter(data1, data2)
plt.show()

# summarize
print('data1: mean=%.3f stdv=%.3f' % (mean(data1), std(data1)))
print('data2: mean=%.3f stdv=%.3f' % (mean(data2), std(data2)))
print('Covariance: %.3f' % covariance[0][1])
print('Pearsons correlation: %.3f' % corr)
```

## Pearson's Correlation and Covariance

This section generates two random datasets (`data1` and `data2`) using a uniform distribution. It calculates:

- The **covariance matrix** to measure how the variables change together.
- The **Pearson correlation coefficient** to measure the strength and direction of their linear relationship.

A scatter plot visualizes the relationship, and summary statistics (mean, standard deviation, covariance, correlation) are printed.


```python
import matplotlib.pyplot as plt
from scipy import stats

#Create the arrays that represent the values of the x and y axis
x = [5,7,8,7,2,17,2,9,4,11,12,9,6]
y = [99,86,87,88,111,86,103,87,94,78,77,85,86]

#Execute a method that returns some important key values of Linear Regression
slope, intercept, r, p, std_err = stats.linregress(x, y)

# measure the correlation 
corr, _ = stats.pearsonr(x, y)
print('Pearsons correlation: %.3f' % corr)

#Create a function that uses the slope and intercept values to return a new value. 
#This new value represents where on the y-axis the corresponding x value will be placed
def myfunc(x):
  return slope * x + intercept

#Run each value of the x array through the function. This will result in a new array with new values for the y-axis
mymodel = list(map(myfunc, x))

#Draw the original scatter plot & the line of linear regression
plt.scatter(x, y)
plt.plot(x, mymodel)
plt.show()
```

## Simple Linear Regression with Fixed Data

Here, we define two lists `x` and `y` representing paired observations. Using `scipy.stats.linregress`, we compute:

- `slope` and `intercept` of the best-fit line
- `r` (correlation coefficient), `p` (p-value), and `std_err` (standard error)

We define a function `myfunc` to predict `y` values from `x`, then plot both the original data and the regression line.

```python
import matplotlib.pyplot as plt
from scipy import stats
import random

#Create the arrays that represent the values of the x and y axis randomly 
x= [random.randint(1, 20 ) for _ in range(18)]
y = [random.randint(6, 100 ) for _ in range(18)]
#Execute a method that returns some important key values of Linear Regression
slope, intercept, r, p, std_err = stats.linregress(x, y)

# measure the correlation 
corr, _ = stats.pearsonr(x, y)
print('Pearsons correlation: %.3f' % corr)

#Create a function that uses the slope and intercept values to return a new value. 
#This new value represents where on the y-axis the corresponding x value will be placed
def myfunc(x):
  return slope * x + intercept

#Run each value of the x array through the function. This will result in a new array with new values for the y-axis
mymodel = list(map(myfunc, x))

#Draw the original scatter plot & the line of linear regression
plt.scatter(x, y)
plt.plot(x, mymodel)
plt.show()
```
